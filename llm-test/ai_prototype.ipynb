### Models Less Than 1B: (Pengyi)

##### Model 1

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("prithivMLmods/FastThink-0.5B-Tiny")
model = AutoModelForCausalLM.from_pretrained("prithivMLmods/FastThink-0.5B-Tiny")

prompts = [
    "What are the apartments that allows cat in champaign?",
    "hey yo gimme a list of places that are lower than 400 per month i'm broke af",
    "meow?",
]

for i, prompt in enumerate(prompts, 1):
    print(f"--- Response to Prompt {i} ---")

    messages = [
        {"role": "user", "content": prompt},
    ]

    inputs = tokenizer.apply_chat_template(
       messages,
       add_generation_prompt=True,
       tokenize=True,
       return_dict=True,
       return_tensors="pt",
    ).to(model.device)

    outputs = model.generate(**inputs, max_new_tokens=40)

    generated_text = tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:], skip_special_tokens=True)
    print(generated_text)
    print("\n")
```

### Models From 3B to 7B: (Blake)

##### Model 1

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_id = "microsoft/Phi-4-mini-instruct"
tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_id, trust_remote_code=True)

prompts = [
    "What are the apartments that allow cats in Champaign?",
    "hey yo gimme a list of places that are lower than 400 per month i'm broke af",
    "meow?",
]

for i, prompt in enumerate(prompts, 1):
    messages = [{"role": "user", "content": prompt}]
    inputs = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        tokenize=True,
        return_dict=True,
        return_tensors="pt",
    ).to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=80,
        do_sample=False,  # deterministic; set True for variety
    )

    answer_ids = outputs[0][inputs["input_ids"].shape[-1]:]
    answer = tokenizer.decode(answer_ids, skip_special_tokens=True)

    print(f"\n--- Prompt {i} ---\n{prompt}\n\n--- Answer {i} ---\n{answer}")
```

Give three most relevant listings in JSON format and nothing else. Include the primary key of the listing in the database, pricing, size, amenities, and advantage of living there (cheaper? more convenient? less noise?) in the json.

```python
system_prompt = (
    "You are a searchbot for an apartment/sublease aggregation website. The user would inquire about apartment/sublease listings and you would need to output listing details according to user's need. Give three most relevant listings in JSON format and nothing else. Include the primary key of the listing in the database, pricing, size, amenities, and advantage of living there (cheaper? more convenient? less noise?) in the json."
)

prompts = [
    "What are the apartments that allows cat in champaign?",
]

for i, prompt in enumerate(prompts, 1):
    print(f"--- Response to Prompt {i} ---")

    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": prompt},
    ]

    inputs = tokenizer.apply_chat_template(
       messages,
       add_generation_prompt=True,
       tokenize=True,
       return_dict=True,
       return_tensors="pt",
    ).to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=300,
        pad_token_id=tokenizer.eos_token_id
    )

    generated_text = tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:], skip_special_tokens=True)
    print(generated_text)
    print("\n")
```
