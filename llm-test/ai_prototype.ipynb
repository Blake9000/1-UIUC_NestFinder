### Models Less Than 1B: (Pengyi)

##### Model 1

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("prithivMLmods/FastThink-0.5B-Tiny")
model = AutoModelForCausalLM.from_pretrained("prithivMLmods/FastThink-0.5B-Tiny")

prompts = [
    "What are the apartments that allows cat in champaign?",
    "hey yo gimme a list of places that are lower than 400 per month i'm broke af",
    "meow?",
]

for i, prompt in enumerate(prompts, 1):
    print(f"--- Response to Prompt {i} ---")

    messages = [
        {"role": "user", "content": prompt},
    ]

    inputs = tokenizer.apply_chat_template(
       messages,
       add_generation_prompt=True,
       tokenize=True,
       return_dict=True,
       return_tensors="pt",
    ).to(model.device)

    outputs = model.generate(**inputs, max_new_tokens=40)

    generated_text = tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:], skip_special_tokens=True)
    print(generated_text)
    print("\n")
```

### Models From 1B to 3B: (Hector)

##### Model 1

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("LocoreMind/LocoOperator-4B")
model = AutoModelForCausalLM.from_pretrained("LocoreMind/LocoOperator-4B")

prompts = [
    "What apartments in Champaign, IL allow cats?",
    "hey yo gimme a list of places in Champaign under $400 per month i'm broke af",
    "meow?"
]

for prompt in prompts:
    messages = [{"role": "user", "content": prompt}]

    inputs = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        tokenize=True,
        return_dict=True,
        return_tensors="pt",
    ).to(model.device)

    outputs = model.generate(**inputs, max_new_tokens=40)

    print("\n==============================")
    print("PROMPT:", prompt)
    print("RESPONSE:")
    print(tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```

```python
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="gpt2"
)

prompts = [
    "What apartments in Champaign, IL allow cats?",
    "hey yo gimme a list of places in Champaign under $400 per month i'm broke af",
    "meow?"
]

for prompt in prompts:
    print("\n==============================")
    print("PROMPT:", prompt)

    output = pipe(prompt, max_new_tokens=100)
    print("RESPONSE:")
    print(output[0]["generated_text"])
```

### Models From 3B to 7B: (Blake)

##### Model 1

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_id = "microsoft/Phi-4-mini-instruct"
tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_id, trust_remote_code=True)

prompts = [
    "What are the apartments that allow cats in Champaign?",
    "hey yo gimme a list of places that are lower than 400 per month i'm broke af",
    "meow?",
]

for i, prompt in enumerate(prompts, 1):
    messages = [{"role": "user", "content": prompt}]
    inputs = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        tokenize=True,
        return_dict=True,
        return_tensors="pt",
    ).to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=80,
        do_sample=False,  # deterministic; set True for variety
    )

    answer_ids = outputs[0][inputs["input_ids"].shape[-1]:]
    answer = tokenizer.decode(answer_ids, skip_special_tokens=True)

    print(f"\n--- Prompt {i} ---\n{prompt}\n\n--- Answer {i} ---\n{answer}")
```

### Models From 7B to 65B: (Lillian)

##### Model 1

```python
# Load model directly
from transformers import AutoProcessor, AutoModelForImageTextToText

tokenizer = AutoProcessor.from_pretrained("Qwen/Qwen3-VL-8B-Instruct")
model = AutoModelForImageTextToText.from_pretrained("Qwen/Qwen3-VL-8B-Instruct")

prompts = [
    "What are the apartments that allows cat in champaign?",
    "hey yo gimme a list of places that are lower than 400 per month i'm broke af",
    "meow?",
]

for i, prompt in enumerate(prompts, 1):
    print(f"--- Response to Prompt {i} ---")

    messages = [
        {
            "role": "user",
            "content": [{"type": "text", "text": prompt}]
        }
    ]

    inputs = tokenizer.apply_chat_template(
       messages,
       add_generation_prompt=True,
       tokenize=True,
       return_dict=True,
       return_tensors="pt",
    ).to(model.device)

    outputs = model.generate(**inputs, max_new_tokens=40)

    # Decode the output
    generated_text = tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:], skip_special_tokens=True)
    print(generated_text)
    print("\n")
```

##### Model 3

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True
)

model_id = "vhab10/llama-3-8b-merged-linear"
tokenizer = AutoTokenizer.from_pretrained(model_id)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map={"": "cuda:0"},
    quantization_config=quant_config
)

prompts = [
    "What are the apartments that allows cat in champaign?",
    "hey yo gimme a list of places that are lower than 400 per month i'm broke af",
    "meow?",
]

for i, prompt in enumerate(prompts, 1):
    print(f"--- Response to Prompt {i} ---")

    messages = [
        {"role": "user", "content": prompt},
    ]

    inputs = tokenizer.apply_chat_template(
       messages,
       add_generation_prompt=True,
       tokenize=True,
       return_dict=True,
       return_tensors="pt",
    ).to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=150,
        pad_token_id=tokenizer.eos_token_id
    )

    generated_text = tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:], skip_special_tokens=True)
    print(generated_text)
    print("\n")
```
